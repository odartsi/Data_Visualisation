# Data Visualisation with matplotlib and seaborn

## Part 1 - The basics


Matplotlib and Seaborn are both Python visualization libraries, but they have some differences in terms of functionality, aesthetics, and ease of use. Here are some key differences between the two:

Functionality: Matplotlib is a low-level library that provides a wide range of customization options for creating various types of plots. Seaborn, on the other hand, is built on top of Matplotlib and provides a higher-level interface for creating some specific types of plots that are commonly used in statistical data analysis, such as heat maps, violin plots, and factor plots.

Aesthetics: Seaborn has a more modern and aesthetically pleasing default style than Matplotlib, with better color palettes and more sophisticated plot elements. Seaborn also provides built-in themes that allow you to quickly change the overall look and feel of your plots.

Ease of use: Seaborn is generally considered to be easier to use than Matplotlib for common statistical visualizations, as it provides a more concise and intuitive syntax for creating such plots. Seaborn also includes some convenience functions for performing statistical analysis and data manipulation in conjunction with plotting.

Customization: While Seaborn is more limited in terms of the types of plots it can create compared to Matplotlib, it provides a range of built-in options for customizing plot elements such as colors, fonts, labels, and axes. Matplotlib provides more extensive customization options, but at the cost of being more verbose and requiring more effort to produce a polished plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the tips dataset from Seaborn
tips = sns.load_dataset('tips')

# Set the figure size
sns.set(rc={'figure.figsize':(8,6)})

# Set the color palette
sns.set_palette('deep')

# Create the scatter plot
ax = sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)

# Set the title and axes labels
ax.set_title('Total Bill vs. Tip by Gender')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

# Add a legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Female', 'Male'])

# Show the plot
sns.despine()
sns.set_style("whitegrid")
sns.set_context("talk")
sns.despine()
plt.show()
```

Certainly! Here's an overview of what each of these Seaborn functions does:

1. **sns.despine**: The `sns.despine` function is used to remove the top and right spines from a plot, which are the lines that border the right and top edges of the plot area. By default, `sns.despine` removes both spines, but you can specify which spines to remove using the `left`, `right`, `top`, and `bottom` arguments. For example, `sns.despine(left=True)` would remove only the left spine.

2. **sns.set_style**: The `sns.set_style` function is used to set the visual style of a plot. Seaborn provides several built-in styles, including `darkgrid`, `whitegrid`, `dark`, `white`, and `ticks`. These styles adjust the color palette, font size, grid lines, and other visual elements of the plot. For example, `sns.set_style('darkgrid')` would set the plot style to use a dark background with grid lines.

3. **sns.set_context**: The `sns.set_context` function is used to set the context of a plot, which adjusts the scale and font size of the plot elements. Seaborn provides four contexts: `paper`, `notebook`, `talk`, and `poster`, which correspond to increasing sizes and font scales. For example, `sns.set_context('poster')` would set the plot to use the largest font size and scale for a poster-sized plot.

All three of these functions are optional and can be used to customize the style and appearance of your Seaborn plots. By default, Seaborn provides sensible defaults for these settings, but you can modify them to suit your preferences or specific use cases.

## Pie plots

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create some data for the pie chart
sizes = [15, 30, 45, 10]
labels = ['Group A', 'Group B', 'Group C', 'Group D']

# Create the pie chart using Matplotlib
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)

# Add a title
ax.set_title('Distribution of Groups')

# Use Seaborn to style the plot
sns.set_style('white')

# Show the plot
plt.show()
```

```python
# define Seaborn color palette to use
palette_color = sns.color_palette('dark')
# declaring exploding pie

fig, ax = plt.subplots()
explode = [0, 0.1, 0, 0]
ax.pie(sizes,colors=palette_color,explode=explode,labels=labels, autopct='%.0f%%')

# Add a title
ax.set_title('Distribution of Groups')

# Use Seaborn to style the plot
sns.set_style('white')

# Show the plot
plt.show()
```

A line plot is used to show the trend of a variable over time or another continuous range of values. In a line plot, the x-axis represents the range of values for the variable, while the y-axis represents the values of the variable itself. A line is then drawn connecting the points on the plot that represent the values of the variable at different points in time or the range of values.

On the other hand, a histogram is used to show the distribution of a variable across a range of values. In a histogram, the x-axis represents the range of values for the variable, while the y-axis represents the frequency or count of observations that fall within each bin or range of values. The data is split into bins, and the height of each bar in the histogram represents the number of observations that fall within that bin.

So the main difference between line plots and histograms is the type of data they are used to visualize. Line plots are used for continuous data that can be measured over a range of values, while histograms are used to show the distribution of data across a range of values.

```python
import panel as pn
pn.extension('tabulator')
import hvplot.pandas
import seaborn as sns
import numpy as np
import pandas as pd


tips = sns.load_dataset("tips")

# Define panel widgets
bill_slider=pn.widgets.IntSlider(name='total_bill', start=0,end=int(max(tips.total_bill)),step=1, value=10)
bill_slider


tips=tips.interactive() ## super important

# Create radio buttons
y_axis_tip=pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['tip'],
    button_type='success')

sex=['Female','Male']

#create the pipeline
tip_pipeline=(
    tips[(tips.total_bill<=bill_slider) & (tips.sex.isin(sex))]
    .groupby(['sex','total_bill'])[y_axis_tip].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='total_bill')
    .reset_index(drop=True)
)

tip_plot=tip_pipeline.hvplot(x='total_bill',by='sex',y=y_axis_tip, title='bill vs tips')
```
